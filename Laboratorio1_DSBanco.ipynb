```python
import pandas as pd
```

```python
ds_banco = pd.read_csv("/content/dataset_banco.csv")
```

```python
ds_banco
```

```python
ds_banco.info()
```

```python
# Imputación de valores nulos con la moda para las columnas categóricas 'job', 'marital' y 'education'
ds_banco['job'].fillna(ds_banco['job'].mode()[0], inplace=True)
ds_banco['marital'].fillna(ds_banco['marital'].mode()[0], inplace=True)
ds_banco['education'].fillna(ds_banco['education'].mode()[0], inplace=True)
```

```python
ds_banco.info()
```

```python
# Imputación de valores nulos con la mediana para las columnas numéricas 'balance' y 'duration'
ds_banco['balance'].fillna(ds_banco['balance'].median(), inplace=True)
ds_banco['duration'].fillna(ds_banco['duration'].median(), inplace=True)
```

```python
ds_banco.info()
```

```python
# Imputación de valores nulos en 'pdays' con -1
ds_banco['pdays'].fillna(-1, inplace=True)
```

```python
ds_banco.info()
```

```python
null_counts_imputation = ds_banco.isnull().sum()
print(null_counts_imputation)
```

```python
# Descripción general para identificar posibles outliers
ds_banco.describe()
```

```python
from scipy import stats
import numpy as np

# Usando Z-score para detectar outliers
z_scores = stats.zscore(ds_banco['balance'])
outliers = ds_banco[np.abs(z_scores) > 3]

print(outliers)
```

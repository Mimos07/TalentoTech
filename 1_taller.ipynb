```python
import pandas as pd
```

```python
leerArchivo = pd.read_excel("/content/example_ventas.xlsx")
```

```python
print("Ejercicio 1: Mostrar las primeras filas del DataFrame")
leerArchivo.head(5)
```

```python
print("Ejercicio2: Ventas por Región")
ventas_por_region= leerArchivo.groupby('Region')['Ventas'].sum()
print(ventas_por_region)
print("\n")
```

```python
print("Ejercicio 3: Filtrar el DataFrame para mostrar solo las ventas realizadas por David")
ventas_david = leerArchivo[leerArchivo ['Vendedor' ] == 'David']
print(ventas_david)
```

```python
print("Ejercicio 4: Calcular el promedio de ventas por region")
ventas_por_producto = leerArchivo.groupby ('Region' )['Ventas'].mean()
print(ventas_por_producto)
```

```python
print("Ejercicio 5: Calcular el promedio de ventas por region con dos decimales")
ventas_por_producto = leerArchivo.groupby('Region')['Ventas'].mean().round(2)
print(ventas_por_producto)
```

```python
print("Ejercicio 6: Ordenar el DataFrame por ventas en orden descendente")
df_ordenado = leerArchivo.sort_values(by='Ventas', ascending=False)
print (df_ordenado)
```

```python
print("Ejercicio 7: Filtrar las filas con ventas mayores a 10000")
df_filtrado = leerArchivo[leerArchivo ['Ventas']>10000].sort_values (by='Ventas', ascending=True)
print(df_filtrado)
print ("\n")
```

```python
print("Ejercicio 9: Eliminar la columna'Vendedor'")
df_sin_vendedor = leerArchivo.drop(columns=[ 'Vendedor' ])
print(df_sin_vendedor)
print("\n")
```
